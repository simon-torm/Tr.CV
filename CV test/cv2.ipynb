```python
color1 = (255, 50, 50)
color2 = (0, 100, 255)

img = np.zeros((512, 512, 3), np.uint8) + 50

cv.line(img, (0, 0), (511, 511), color1, 3)
cv.rectangle(img, (200, 100), (300, 200), color2, 5)
cv.circle(img, (300, 400), 50, color1, -1)

coords = np.array([[100, 200], [400, 250], [500, 350], [50, 320]])
cv.polylines(img, [coords], True, color2, 5)

font = cv.FONT_HERSHEY_PLAIN
cv.putText(img, 'Hello world!', (100, 400), font, 3, (240, 240, 240), 3)

plt.imshow(img)
```
